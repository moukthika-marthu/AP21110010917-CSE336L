### 2. Import the Iris dataset. Write a program to obtain the Euclidian Distance Matrix for all the data samples in the feature space. Distance metric is a 2D array, where the (i,j)th entry represents the distance between the ith and jth sample points in the feature space.

```python
from sklearn.datasets import load_iris
import numpy as np

# Load the Iris dataset
iris = load_iris()
data = iris.data

# Function to calculate Euclidean distance between two points
def euclidean_distance(point1, point2):
    return np.linalg.norm(point2 - point1)

# Function to calculate Euclidean Distance Matrix for all samples
def euclidean_distance_matrix(data):
    num_samples = len(data)
    distance_matrix = np.zeros((num_samples, num_samples))

    for i in range(num_samples):
        for j in range(i, num_samples):
            distance = euclidean_distance(data[i], data[j])
            distance_matrix[i][j] = distance
            distance_matrix[j][i] = distance  # Distance matrix is symmetric

    return distance_matrix

# Calculate Euclidean Distance Matrix
distance_matrix = euclidean_distance_matrix(data)

# Print the distance matrix
print("Euclidean Distance Matrix:")
print(distance_matrix)
```

### 3. Import the Iris dataset. Prepare a dataset considering samples belong to any two output classes. Draw the scatter plot for all th samples in the new dataset considering any two input attributes. Examine the scatter plot to find the equation of a line that  can separate sample of two classes.

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import numpy as np

# Load the Iris dataset
iris = load_iris()

# Extracting input features (attributes) and output labels
X = iris.data
y = iris.target

# Selecting samples belonging to two output classes (e.g., classes 0 and 1)
class_indices = [0, 1]  # Selecting classes 0 and 1
X_subset = X[np.isin(y, class_indices)]
y_subset = y[np.isin(y, class_indices)]

# Selecting two input attributes for scatter plot (e.g., sepal length and sepal width)
attribute1_index = 0  # Sepal length
attribute2_index = 1  # Sepal width

# Scatter plot for the selected input attributes
plt.figure(figsize=(8, 6))
plt.scatter(X_subset[y_subset == class_indices[0], attribute1_index],
            X_subset[y_subset == class_indices[0], attribute2_index],
            label=f'Class {class_indices[0]}', c='blue')
plt.scatter(X_subset[y_subset == class_indices[1], attribute1_index],
            X_subset[y_subset == class_indices[1], attribute2_index],
            label=f'Class {class_indices[1]}', c='red')

plt.xlabel(iris.feature_names[attribute1_index])
plt.ylabel(iris.feature_names[attribute2_index])
plt.title('Scatter Plot of Two Output Classes')
plt.legend()
plt.grid(True)
plt.show()
```
